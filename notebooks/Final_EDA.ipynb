# Data Imputing


## Subset

```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.tree import DecisionTreeRegressor

# Load
df = pd.read_parquet("../data/subset.parquet")

# 1. Drop all INDEX_ columns
index_cols = [col for col in df.columns if col.startswith('INDEX_')]
df = df.drop(columns=index_cols)

# 2. Build the list of columns to keep:
#    – all numeric columns
#    – plus 'country_text_id' no matter what
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cols_to_keep = list(dict.fromkeys(numeric_cols + ['country_text_id']))  # dict.fromkeys preserves order and dedups

# 3. Subset
df = df[cols_to_keep]
df
```

```python
# 1. Set up the imputer once
imputer = IterativeImputer(
    estimator=DecisionTreeRegressor(random_state=0),
    max_iter=10,
    random_state=0
)

imputed_dfs = []

for country, group in df.groupby('country_text_id', sort=False):
    # 2. Sort by year
    grp = group.sort_values('year')
    
    # 3. Save identifiers
    years = grp['year'].values
    country_ids = grp['country_text_id'].values
    
    # 4. Build X: drop identifiers, keep numeric only
    X = (
        grp
        .drop(columns=['country_text_id', 'year'])
        .select_dtypes(include=[np.number])
    )
    
    # 5. ***Drop any column that’s all‐NA in this country***
    X = X.dropna(axis=1, how='all')
    
    # 6. Impute
    X_imputed = imputer.fit_transform(X)
    
    # 7. Rebuild DataFrame
    df_imp = pd.DataFrame(
        X_imputed,
        columns=X.columns,
        index=grp.index
    )
    # 8. Reattach identifiers
    df_imp['country_text_id'] = country_ids
    df_imp['year'] = years
    
    imputed_dfs.append(df_imp)

# 9. Concatenate back and restore order
df_filled = pd.concat(imputed_dfs).sort_index()

print(df_filled.head())
```

# Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.spatial import ConvexHull
import matplotlib.colors as mcolors
from matplotlib.patches import Ellipse

# Filter for the desired year (e.g., 2024)
df_2024 = df[df['year'] == 2024].copy()

# Define ID columns and features
id_columns = ['country_name', 'year']
feature_cols = [col for col in df_2024.columns if col not in id_columns]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_2024[feature_cols])

# PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
pca_df = pd.concat([pca_df, df_2024[id_columns].reset_index(drop=True)], axis=1)

# Clustering
n_clusters = min(6, len(pca_df))
if n_clusters > 2:
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    pca_df['Cluster'] = kmeans.fit_predict(pca_df[['PC1', 'PC2']])
else:
    pca_df['Cluster'] = 0

# Visualization
fig, ax = plt.subplots(figsize=(16, 10), facecolor='white')
ax.set_facecolor('white')
colors = list(mcolors.TABLEAU_COLORS.values()) + list(mcolors.CSS4_COLORS.values())
cmap = {i: colors[i] for i in range(n_clusters)}

def plot_convex(points, ax, color, alpha=0.1):
    if len(points) < 3:
        return
    try:
        hull = ConvexHull(points)
        ax.fill(points[hull.vertices, 0], points[hull.vertices, 1], color=color, alpha=alpha)
    except:
        x_mean, y_mean = np.mean(points[:, 0]), np.mean(points[:, 1])
        x_std, y_std = np.std(points[:, 0]), np.std(points[:, 1])
        ellipse = Ellipse((x_mean, y_mean), width=4*x_std, height=4*y_std,
                          fill=True, color=color, alpha=alpha)
        ax.add_patch(ellipse)

# Plot hulls and points
for cluster in range(n_clusters):
    data = pca_df[pca_df['Cluster'] == cluster]
    points = data[['PC1', 'PC2']].values
    if len(points) >= 3:
        plot_convex(points, ax, cmap[cluster])
    plt.scatter(data['PC1'], data['PC2'], c=[cmap[cluster]] * len(data), s=50, alpha=0.8)

# Label positions
positions = [(row['PC1'] + 0.1, row['PC2']) for _, row in pca_df.iterrows()]

def adjust_positions(pos, iterations=50):
    adj = pos.copy()
    for _ in range(iterations):
        for i in range(len(adj)):
            for j in range(len(adj)):
                if i != j:
                    dx, dy = adj[i][0] - adj[j][0], adj[i][1] - adj[j][1]
                    dist = np.hypot(dx, dy)
                    if dist < 0.5:
                        dx, dy = dx / (dist + 1e-6), dy / (dist + 1e-6)
                        adj[i] = (adj[i][0] + 0.1 * dx, adj[i][1] + 0.1 * dy)
    return adj

adjusted = adjust_positions(positions)

for i, row in pca_df.iterrows():
    plt.annotate(row['country_name'], (row['PC1'], row['PC2']),
                 xytext=(adjusted[i][0] - row['PC1'], adjusted[i][1] - row['PC2']),
                 textcoords='offset points', fontsize=9, ha='center', alpha=0.6)

# Axis labels with variance
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
plt.title('V-Dem Country Clusters in PCA Space (2024)')

# Add grid and reference lines
plt.grid(alpha=0.3, linestyle='--')
plt.axhline(0, color='gray', lw=1, alpha=0.3)
plt.axvline(0, color='gray', lw=1, alpha=0.3)

# Interpretation guide
pc1_load = pd.Series(pca.components_[0], index=feature_cols)
pc2_load = pd.Series(pca.components_[1], index=feature_cols)

guide_text = ["INTERPRETATION GUIDE", "-" * 25]
guide_text += ["PC1 →"]
guide_text += [f"  + {k.replace('_',' ').title()} ({v:.2f})" for k, v in pc1_load.sort_values(ascending=False).head(3).items()]
guide_text += [f"  - {k.replace('_',' ').title()} ({v:.2f})" for k, v in pc1_load.sort_values().head(3).items()]
guide_text += ["", "PC2 ↑"]
guide_text += [f"  + {k.replace('_',' ').title()} ({v:.2f})" for k, v in pc2_load.sort_values(ascending=False).head(3).items()]
guide_text += [f"  - {k.replace('_',' ').title()} ({v:.2f})" for k, v in pc2_load.sort_values().head(3).items()]

plt.figtext(0.78, 0.5, "\n".join(guide_text), fontsize=9,
            va='center', bbox=dict(boxstyle='round', facecolor='white', edgecolor='lightgray', alpha=0.9))

plt.tight_layout(rect=[0, 0, 0.75, 1])
plt.savefig("vdem_clusters_2024.png", dpi=300, facecolor='white')
plt.show()

# Summary
print("Top contributors to PC1:")
print(pc1_load.abs().sort_values(ascending=False).head(10))
print("\nTop contributors to PC2:")
print(pc2_load.abs().sort_values(ascending=False).head(10))

if n_clusters > 1:
    print("\nCluster breakdown:")
    for c in range(n_clusters):
        countries = pca_df[pca_df['Cluster'] == c]['country_name'].tolist()
        print(f"Cluster {c+1}: {', '.join(countries)}")
```
